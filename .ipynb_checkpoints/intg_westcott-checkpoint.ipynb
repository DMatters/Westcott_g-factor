```python
# Import packages
import math as m
import numpy as np
from scipy.integrate import quad, trapezoid, simpson
import matplotlib.pyplot as plt
import csv
```

```python
# Define constants
kB = 1.38066e-23 #J/K
h = 6.62618e-34 #J*s
m_n = 1.00866501 *1.660566e-27 #kg
v_0 = 2200 #m/s
E_0_J = 1/2 * m_n * v_0**2 #J
E_0 = E_0_J / 1.602189e-19 #eV
```

```python
# Test case: 197Au
# Resonance parameters
E_r = 4.906 #eV
Gamma = 137.7/1000 #eV
```

```python
# Define functions
# Lorentzian resonance; Molnar eqn. 1-3
def del_0(v):
    E = 1/2 * m_n * v**2 / 1.602189e-19 #eV
    return ((E_r - E_0)**2 + Gamma**2/4)/((E_r - E)**2 + Gamma**2/4)

# Neutron density, given Maxwellian distribution; Molnar Table 1-1
def p(T,v):
    v_T = m.sqrt(2*kB*T/m_n)
    return 4/m.sqrt(m.pi) * m.exp(-v**2/v_T**2) * v**2/v_T**3 
```

```python
# These differ slightly from the values in the IAEA Database (there's no value in Molnar), but the behavior is similar
# Next to try the cold neutron source case...first need to import the neutron density function
with open('cold_spectrum.csv',mode='r', encoding='utf-8-sig') as file:
    csvFile = csv.reader(file)
    En = []
    dndE = []
    for lines in csvFile:
        En.append(float(lines[0]))
        dndE.append(float(lines[1]))

# Plot to ensure it looks correct
plt.plot(En,dndE)
ax = plt.gca()
ax.set_xlim([0,0.1])
#ax.set_yscale('log')
plt.show()
```

```python
# Perform integration over velocity to find Westcott g-factor
# Westcott g-factor; Molnar eqn. 1-5
def g(T):
    v = []
    for i in range(len(En)):
        v.append(m.sqrt(2*En[i]*1.602e-19/m_n))
    g = 0
    for i in range(len(v)):
        if i==0:
            g += del_0(v[i])*p(T,v[i])*(v[0] - 0)
        else:
            g += del_0(v[i])*p(T,v[i])*(v[i] - v[i-1])
    return g

# Now evaluate the g-factors
for t in T_list:
    print(t, 'K      g =', g(t))
```
